```python
img = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/WindowsLogo.jpg")
```

```python
plt.imshow(img[:, :, ::-1])
```

```python
img.shape
```

```python
histograma_b = cv.calcHist([img], [0], None, [256], (0, 256))
plt.plot(histograma_b, "b")
plt.show()
```

```python
histograma_g = cv.calcHist([img], [1], None, [256], (0, 256))
plt.plot(histograma_b, "g")
plt.show()
```

```python
histograma_r = cv.calcHist([img], [2], None, [256], (0, 256))
plt.plot(histograma_b, "r")
plt.show()
```

```python
colores = ["b", "g", "r"]
```

```python
for n, i in enumerate(colores):
    tmp = cv.calcHist([img], [n], None, [256], (0, 256))
    plt.plot(tmp, i)
plt.show()
```

```python
for cuenta, valor in enumerate(colores):
    print(f"Cuenta = {cuenta} | Valor = {valor}") # f-string
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
sudoku = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/sudoku.png", 0)
```

```python
plt.imshow(sudoku, "gray")
plt.axis("off")
plt.show()
```

---
## Umbralización Global

```python
# Binarizar
sudoku_bin_tradicional = cv.threshold(sudoku, 50, 255, cv.THRESH_BINARY_INV)[1]
```

```python
plt.imshow(sudoku_bin_tradicional, "gray")
plt.axis("off")
plt.show()
```

---
## Umbralización Adaptativa

```python
# Promedio
blocksize = 11
C = 2
sudoku_bin_adaptive = cv.adaptiveThreshold(sudoku, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, blocksize, C)
```

```python
plt.imshow(sudoku_bin_adaptive, "gray")
plt.axis("off")
plt.title(f"Blocksize={blocksize} y C={C}")
plt.show()
```

```python
# Gaussiana
blocksize = 17
C = 2
sudoku_bin_adaptive = cv.adaptiveThreshold(sudoku, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, blocksize, C)

plt.imshow(sudoku_bin_adaptive, "gray")
plt.axis("off")
plt.title(f"Blocksize={blocksize} y C={C}")
plt.show()
```

```python
# Umbral global, recomendado OTSU
plt.imshow(cv.threshold(sudoku, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)[1], "gray")
plt.show()
```

```python
babuino = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/baboon.jpg", 0)
```

```python
# Umbral global, recomendado OTSU
th, img = cv.threshold(babuino, None, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)
plt.imshow(img, "gray")
plt.title(f"Imagen con umbral {th}")
plt.show()
```

```python
# Umbral global, recomendado TRIANGLE
th, img = cv.threshold(babuino, None, 255, cv.THRESH_BINARY + cv.THRESH_TRIANGLE)
plt.imshow(img, "gray")
plt.title(f"Imagen con umbral {th}")
plt.show()
```

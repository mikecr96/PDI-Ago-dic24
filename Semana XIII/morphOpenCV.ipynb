```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
j = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/j.png", 0)
```

```python
np.unique(j)
```

---
Erode

```python
assert j is not None, "file could not be read, check with os.path.exists()"
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(j,kernel,iterations = 1)
plt.imshow(erosion, "gray")
plt.show()
```

---
Dilate

```python
dilation = cv.dilate(j,kernel,iterations = 1)
plt.imshow(dilation, "gray")
plt.show()
```

---
Opening

```python
apertura = cv.morphologyEx(j, cv.MORPH_OPEN, kernel)
plt.imshow(apertura, "gray")
plt.show()
```

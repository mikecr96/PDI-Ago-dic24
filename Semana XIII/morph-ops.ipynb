```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_gray = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/pic3.png", 0)
```

---
binarizaci√≥n

```python
img_bin = cv.threshold(img_gray, 180, 255, cv.THRESH_BINARY_INV)[1]
```

```python
plt.imshow(img_bin, "gray")
plt.show()
```

```python
np.unique(img_bin)
```

```python
se = np.ones((5,5) , np.uint8)
```

---
Erosi√≥n

```python
# plt.imshow(cv.erode(cv.erode(img_bin, se), se), "gray")
plt.imshow(cv.erode(img_bin, se, iterations=2), "gray")
plt.show()
```

---
Dilataci√≥n

```python
plt.imshow(cv.dilate(img_bin, se, iterations=10), "gray", vmin=0)
plt.show()
```

---
BONUS

```python
img_bin = cv.threshold(img_gray, 180, 255, cv.THRESH_BINARY)[1]
plt.imshow(cv.erode(img_bin, se, iterations=10), "gray", vmin=0)
plt.show()
```

---
Contornos

```python
"""Detecci√≥n de bordes: tomando la dilataci√≥n de una imagen y
luego restando la imagen original, se resalta as√≠ solo esos
nuevos p√≠xeles en los bordes de los objetos que fueron
agregados por la dilataci√≥n."""

dilatada = cv.dilate(img_bin, se)

plt.imshow(dilatada - img_bin, "gray")
plt.show()
```

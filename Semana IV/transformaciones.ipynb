# Negativo

```python
import numpy as np
```

```python
img_neg = np.zeros(img.shape, dtype=np.uint8)
```

```python
img_neg.dtype
```

```python
img.shape[0]
```

```python
img.size
```

```python
for f in range(img.shape[0]): # Filas
    for c in range(img.shape[1]): # Columnas
        img_neg[f, c] = 255 - img[f, c]
```

```python
plt.imshow(img_neg, cmap="gray")
plt.show()
```

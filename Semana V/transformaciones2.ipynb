# Transformación umbralización

## Método 1. Iterando en todos los píxeles

Esta transformada sirve para binarizar una imagen

```python
img_umb = np.zeros(img.shape, dtype=np.uint8)
```

```python
img.shape
```

```python
fil, col = img.shape
```

```python
k = 200

for f in range(fil): # Filas
    for c in range(col): # Columnas
        if img[f, c] < k:
            img_umb[f, c] = 0
        else:
            img_umb[f, c] = 255

plt.imshow(img_umb, "gray")
plt.show()
```

```python
img_umb2 = cv.threshold(img, k, 255, cv.THRESH_BINARY)[1] # cv.THRESH_BINARY = 0
```

```python
plt.imshow(img_umb2, "gray")
plt.show()
```

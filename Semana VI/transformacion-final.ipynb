```python
import math
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

# Transformación logarítmica

```python
img = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/aero3.jpg", 0)
```

```python
plt.imshow(img, "gray")
plt.show()
```

```python
m = np.max(img)
```

```python
C = 255 / (math.log10(1 + int(m)))
```

```python
C
```

```python
fil, col = img.shape
```

```python
img_log = np.zeros(img.shape, dtype=np.uint8)
```

```python
for f in range(fil): # Filas
    for c in range(col): # Columnas
        img_log[f, c] = C * math.log10(1 + img[f, c])
```

```python
img.dtype
```

```python
img_log.dtype
```

```python
np.hstack((img, img_log))
```

```python
plt.imshow(np.hstack((img, img_log)), "gray")
plt.show()
```

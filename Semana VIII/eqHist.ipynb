Eq. Hist en imgenes a color en formato RGB
##### NO SE HACE

```python
chicky_color = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/chicky_512.png")[:, :, ::-1]
plt.imshow(chicky_color)
plt.show()
```

```python
"""
Pasos para ecualizar RGB

1) Dividir mis canales
2) Ecualizar cada canal
3) Unir los canales YA ECUALIZADOS
4) Mostrar...
"""
```

```python
# 1)
canal_R = chicky_color[:, :, 0]
canal_G = chicky_color[:, :, 1]
canal_B = chicky_color[:, :, 2]
```

```python
# 1)
canal_R, canal_G, canal_B = cv.split(chicky_color)
```

```python
# 2)
canal_R_eq = cv.equalizeHist(canal_R)
canal_G_eq = cv.equalizeHist(canal_G)
canal_B_eq = cv.equalizeHist(canal_B)
```

```python
# 3)
chicky_color_eq = cv.merge((canal_R_eq, canal_G_eq, canal_B_eq))
```

```python
# 4)
plt.imshow(chicky_color_eq)
plt.show()
```

Mejorar el contraste en el espacio HSV - HSB

Pasos para ecualizar RGB

1) Convertir de RGB a HSV
2) Dividir mis canales
3) Ecualizar el canal V (Value, guarda la info. de brillo) 
4) Unir los canales YA ECUALIZADOS -> HSV
5) Regreso de HSV a RGB
6) Mostrar...


```python
PI = 3.14151926
```

```python
cv.COLOR_RGB2HSV
```

```python
ballena_hsv = cv.cvtColor(ballena, cv.COLOR_RGB2HSV)
```

```python
# No mostramos imgenes en HSV (en matplotlib)
# plt.imshow(ballena_hsv)
```

```python
h, s, v = cv.split(ballena_hsv)
```

```python
v_eq = cv.equalizeHist(v)
```

```python
ballena_hsv_eq = cv.merge((h, s, v_eq))
```

```python
plt.imshow(ballena_hsv_eq)
plt.show()
```

```python
ballena_rgb_eq = cv.cvtColor(ballena_hsv_eq, cv.COLOR_HSV2RGB)
```

```python
plt.imshow(np.hstack((ballena, ballena_rgb_eq)))
plt.show()
```

Mejorar el contraste en el espacio YCbCr - Y guarda la luminancia | CbCr son crominancias

Pasos para ecualizar RGB

1) Convertir de RGB a YCbCr
2) Dividir mis canales
3) Ecualizar el canal Y (Y, guarda la info. de brillo) 
4) Unir los canales YA ECUALIZADOS -> YCbCr
5) Regreso de YCbCr a RGB
6) Mostrar...



```python
chicky_ycbcr = cv.cvtColor(chicky_color, cv.COLOR_RGB2YCrCb)
```

```python
# plt.imshow(chicky_ycbcr); plt.imshow()
```

```python
y, cr, cb = cv.split(chicky_ycbcr)
```

```python
y_eq = cv.equalizeHist(y)
```

```python
chicky_ycbcr_eq = cv.merge((y_eq, cr, cb))
```

```python
chicky_rgb_eq = cv.cvtColor(chicky_ycbcr_eq, cv.COLOR_YCrCb2RGB)
```

```python
plt.imshow(chicky_rgb_eq)
plt.imshow()
```

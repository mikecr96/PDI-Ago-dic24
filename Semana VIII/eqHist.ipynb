```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
aero3 = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/aero3.jpg", 0)
```

```python
plt.imshow(aero3, "gray")
plt.show()
```

```python
histograma = cv.calcHist([aero3], [0], None, [256], (0, 256))
plt.plot(histograma)
plt.show()
```

```python
# Ecualizacin
aero_eq = cv.equalizeHist(aero3)
```

```python
plt.imshow(np.hstack((aero3, aero_eq)), "gray")
plt.show()
```

```python
histograma_eq = cv.calcHist([aero_eq], [0], None, [256], (0, 256))
plt.plot(histograma_eq)
plt.show()
```

```python
chicky_color = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic24/images/chicky_512.png")[:, :, ::-1]
plt.imshow(chicky_color)
plt.show()
```

```python
"""
Pasos para ecualizar RGB

1) Dividir mis canales
2) Ecualizar cada canal
3) Unir los canales YA ECUALIZADOS
4) Mostrar...
"""
```

```python
# 1)
canal_R = chicky_color[:, :, 0]
canal_G = chicky_color[:, :, 1]
canal_B = chicky_color[:, :, 2]
```

```python
# 1)
canal_R, canal_G, canal_B = cv.split(chicky_color)
```

```python
# 2)
canal_R_eq = cv.equalizeHist(canal_R)
canal_G_eq = cv.equalizeHist(canal_G)
canal_B_eq = cv.equalizeHist(canal_B)
```

```python
# 3)
chicky_color_eq = cv.merge((canal_R_eq, canal_G_eq, canal_B_eq))
```

```python
# 4)
plt.imshow(chicky_color_eq)
plt.show()
```
